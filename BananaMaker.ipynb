```python
# Draw random samples from Salpeter IMF.
# N     ... number of samples.
# alpha ... power-law index.
# M_min ... lower bound of mass interval.
# M_max ... upper bound of mass interval.
def sampleFromSalpeter(N, alpha, M_min, M_max):
    # Convert limits from M to logM.
    log_M_Min = math.log(M_min)
    log_M_Max = math.log(M_max)
    # Since Salpeter SMF decays, maximum likelihood occurs at M_min
    maxlik = math.pow(M_min, 1.0 - alpha)

    # Prepare array for output masses.
    Masses = []
    # Fill in array.
    while (len(Masses) < N):
        # Draw candidate from logM interval.
        logM = random.uniform(log_M_Min,log_M_Max)
        M    = math.exp(logM)
        # Compute likelihood of candidate from Salpeter SMF.
        likelihood = math.pow(M, 1.0 - alpha)
        # Accept randomly.
        u = random.uniform(0.0,maxlik)
        if (u < likelihood):
            Masses.append(M)
    return Masses
```

```python
import numpy as np
import matplotlib.pyplot as plt
def drawMetdist(l,b):
    # Parameters for two Gaussians and their weights
    mu1, sigma1, w1 = -0.35, 0.28, 0.53
    mu2, sigma2, w2 = 0.22, 0.18, 0.46
    n_samples = 10000
    # Determine number of samples for each and combine
    n1 = int(n_samples * w1)
    n2 = int(n_samples * w2)
    samples = np.concatenate([np.random.normal(mu1, sigma1, n1), 
                          np.random.normal(mu2, sigma2, n2)])
    np.random.shuffle(samples)
    # Visualize 
#    plt.hist(samples, bins=50, density=True)
#    plt.show()
    return samples 
```

```python
from scipy.interpolate import interp1d
#import numpy as np

def GetAgeDist(metarray, agearray, metdist):
    # Parameters for two Gaussians and their weights
    # Known data points
    x = np.array(metarray)
    y = np.array(agearray)
    # Create interpolation function (e.g., cubic)
    #f_linear = interp1d(x, y, kind='linear')
    agedist=[]
    f_cubic = interp1d(x, y, kind='cubic')
    for i in range(len(metdist)):
        if ((metdist[i]<metarray[0]) & (metdist[i]>metarray[-1])):
#            print(metdist[i], f_cubic(metdist[i]))
            agedist.append(f_cubic(metdist[i]))
    # Visualize 
#    plt.hist(samples, bins=50, density=True)
#    plt.show()
    return agedist 
```
